```python
import numpy as np 
import pandas as pd 

df = pd.read_csv('TA_Kota_Palembang.csv')
print('Size of weather data frame is :',df.shape)
df.rename(columns = {'Kelembapan':'Kelembaban'}, inplace = True)
df[0:5]
```

```python
import warnings

warnings.filterwarnings('ignore')
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
df.info()
```

```python
df1 = pd.DataFrame ((df),
columns = ['Curah_hujan'])
df.loc [df1['Curah_hujan'] >= 0.5, 'Intensitas Hujan'] = 'Hujan'
df.loc [df1['Curah_hujan'] < 0.5, 'Intensitas Hujan'] = 'Tidak Hujan'
df
```

```python
df['Intensitas Hujan'].value_counts()
```

```python
df['Intensitas Hujan'].replace({'Tidak Hujan': 0, 'Hujan': 1},inplace = True)
df
```

```python
df.info()
```

```python
df = df.drop(columns=['Tahun','Bulan','Hari','Curah_hujan'],axis=1)
df
```

```python
X = df.drop(['Intensitas Hujan'], axis=1)
y = df[['Intensitas Hujan']]
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X, y, test_size=0.25, random_state=42)
X_train.shape, X_test.shape
```

```python
y_train.shape, y_test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
import time

t0=time.time()
clf_dt = DecisionTreeClassifier(criterion='gini', splitter='best', max_depth=5, 
                                min_samples_split=2, min_samples_leaf=1, min_weight_fraction_leaf=0.0, 
                                max_features=None, random_state=None, max_leaf_nodes=None, 
                                min_impurity_decrease=0.0, class_weight=None, ccp_alpha=0.0)
clf_dt.fit(X_train,y_train)
y_pred = clf_dt.predict(X_test)
score = accuracy_score(y_test,y_pred)
print('accuracy :',score)
print('Time taken :' , time.time()-t0)
```

```python
comp_test_pred = y_test.copy()
comp_test_pred['Prediksi Intensitas Hujan'] = y_pred
comp_test_pred.rename(columns = {'Intensitas Hujan':'Aktual', 'Prediksi Intensitas Hujan':'Prediksi'}, inplace = True)
comp_test_pred.sort_index(inplace=True)
comp_test_pred
```
